```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = pd.read_csv('energydata_complete.csv')
df.head()
```

```python
#quiz number 17

X = df[['T2']]
y = df['T6']

lr = LinearRegression()
lr.fit(X, y)
```

```python
y_pred = lr.predict(X)
print(y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))

print(f"{rmse:.3f}")
```

```python
#quiz number 18
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error

df = df.drop(columns=["date", "lights"], errors='ignore')

X = df.drop(columns=["Appliances"])
y = df["Appliances"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
X_test_scaled
```

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
```

```python
y_train_pred = lr.predict(X_train_scaled)
y_train_pred
```

```python
mae_train = mean_absolute_error(y_train, y_train_pred)
print(f"{mae_train:.3f}")
```

```python
#quiz number 19
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))

print(f"{rmse_train:.3f}")
```

```python
#quiz number 20
y_test_pred = lr.predict(X_test_scaled)
mae_test = mean_absolute_error(y_test, y_test_pred)

print(f"{mae_test:.3f}")
```

```python
#quiz number 21
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))

print(f"{rmse_test:.3f}")
```

```python
#quiz number 22

# From the previous output:
# MAE on the training set = 53.742
# RMSE on the training set = 95.216
# MAE on the testing set = 53.643
# RMSE on the testing set = 93.640

# Therefore, No overfitting
```

```python
#quiz number 23

from sklearn.linear_model import Ridge

ridge_model = Ridge()
ridge_model.fit(X_train_scaled, y_train)

y_test_pred_ridge = ridge_model.predict(X_test_scaled)

rmse_test_ridge = np.sqrt(mean_squared_error(y_test, y_test_pred_ridge))

print(f"{rmse_test_ridge:.3f}")
print('yes, there is change')
```

```python
#quiz number 24

from sklearn.linear_model import Lasso

lasso_model = Lasso()
lasso_model.fit(X_train_scaled, y_train)
```
